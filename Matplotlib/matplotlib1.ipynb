```python
#import matplotlib
import matplotlib.pyplot as plt 
```

```python
#see the plots you create in jupyter notebook
%matplotlib inline
```

```python
import numpy as np 
#make a linearly spaced, from 0 to 5 with 11 points
x = np.linspace(0,5,11)
y=x ** 2
```

```python
x
```

```python
y
```

```python
#Functional Method
plt.plot(x,y)
#use plt.show() if you are not in jupyter notebook
```

```python
plt.plot(x,y,'r-')
```

```python
#add labels
plt.plot(x,y)
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.title('Title')
```

```python
#multi plots on same canvus
#plt.subplot(number of rows, number of columns, plot number you are referring to)
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented API Method
#creates a blank canvas
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

#plot on axes
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title Label')
```

```python
#creates an empty figure/canvas
fig = plt.figure()

#add a set of axes to it add_axes(left, bottom, width, height) in relation to the canvas
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5,0.4,0.3])

#add plots to it
axes1.plot(x,y)
axes2.plot(y,x)

#add titles
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
```
