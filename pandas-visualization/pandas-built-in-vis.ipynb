```python
import numpy as np 
import pandas as pd 
%matplotlib inline
import seaborn as sns 
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
#note that this data is a time series (first column shows dates)
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
#histogram of values in column a of df1
df1['A'].hist()
```

```python
#make histogram off of a dataframe of data
df1['A'].plot(kind='hist', bins=30)
```

```python
#instead of kind, you can use hist off of plot
df1['A'].plot.hist(bins=30)
```

```python
#area plot-similar to a normal line plot, but underneath it is colored
df2.plot.area(alpha=0.4)
```

```python
df2.plot.bar()
```

```python
df1.plot.line(x=df1.index,y='B',figsize=(12,3))
```

```python
#create a scatter plot with column A as the x axis and column B on the y axis. Use the s argument to change the size of the values based off of the values of C
df1.plot.scatter(x='A',y='B',s=df1['C']*100)
```

```python
#box plot
df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a', 'b'])
```

```python
#similar to a scatter plot, but they are hexagons
#gridsize is used to increase the size of the shapes. The darker the shape, the most points that are in that area
df.plot.hexbin(x='a', y='b', gridsize=25)
```

```python
#kernel density dataframe
df2['a'].plot.kde()
```
