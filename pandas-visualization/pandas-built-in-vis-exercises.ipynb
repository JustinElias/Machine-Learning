```python
import pandas as pd 
import matplotlib.pyplot as plt 
df3 = pd.read_csv('df3')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

```python
#create a scatterplot with a on the x axis and b on the y axis
#s=size
#c=color
df3.plot.scatter(x='a', y='b',s=50, c='red', figsize=(12,3))
```

```python
#create a histogram of column 'a'
df3['a'].plot.hist()
```

```python
#polish up the histogram plot using ggplot/style sheets
plt.style.use('ggplot')
```

```python
df3['a'].plot.hist(bins=20, alpha =0.5)
```

```python
#create a boxplot comparing the a and b columns
df3[['a','b']].plot.box()
```

```python
#create a kde plot of the column 'd'
df3['d'].plot.kde()
```

```python
#increase line width and make the linstyle dashed
df3['d'].plot.kde(lw=5,ls='--')
```

```python
#create an area plot of all the columns for just the rows up to 30
df3.ix[0:30].plot.area()
```
