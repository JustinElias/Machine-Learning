```python
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()
```

```python
flights.head()
```

```python
#heat maps - primary way to show matrix plot.
#Your data should already be in matrix form before doing a heat map
#matrix form refers to the index name and the column name matching up so that the cell value actually indicates something that is relevant to both names
#variables need to be in both the columns and rows (the 0,1,2,3,4 in the columsn do not count)
tc = tips.corr()
```

```python
sns.heatmap(tc, annot = True, cmap='coolwarm')
```

```python
#the plan now for the flights dataset is to make one of the column variables appear as a row variable
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
#cmap allows you to change the color pallet. line color changes the color of the lines and linewidths changes the spacing between the squares
sns.heatmap(fp, cmap='magma', linecolor='white', linewidths=1)
```

```python
#cluster map - uses hierarchal clustering to produce a clustered version of this heat map
#it tries to cluster columns and rows together based off of their similarity
#for this specific map, the months are out of order because it's trying to cluster similar months together. Same thing goes for the years
#standard_scale groups everything based on color
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```
