```python
import seaborn as sns 
%matplotlib inline 
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np 
```

```python
#Bar plot - based off of mean
#use the estimator argument to change what you want to see. By default, it will show the mean of the values you choose. Change this will estimator
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
#count plot - estimator counts number of occurances, so you only need to set an x value
sns.countplot(x='sex', data=tips)
```

```python
#box plots - shows distribution of categorical data - adding the hue to split up by smoker status. This gives you a whole 'nother layer of data very quickly
sns.boxplot(x ='day', y='total_bill', data=tips, hue='smoker')

#the box plot shows the quartiles of the data set while the whiskers extend to show the rest of the distribution. The dots are the outliers
```

```python
#violen plot - similar to a box plot - shows distribution of data - allows you to plot all components that correspond to actual data points. It is basically showing the KDE
#good for getting more info, but tougher to read/explain. Good for showing to other data scientists
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex',split=True)
```

```python
#strip plot - draws a scatter plot where one variable is categorical
#can't tell how many points are on top of each other
#use the jitter agument to add random noise to spread out the stacked points
#use Hue to add more info
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
#swarm plot - combine the ideas of a strip plot versus a violen plot
#points are adjusted so they don't overlap - gives a better representation of the distribution of values
#not good for large datasets and most people haven't seen them
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
#combine a swarm plot with a violen plot
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
#Factor Plot - takes in x, y, argument. Use the kind argument to tell seaborn what type of chart you want. You can change this to recreate any of the graphs we've done so far
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```
