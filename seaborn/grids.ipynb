```python
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline 
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
#pairplot - automated join plot for the entire dataset
sns.pairplot(iris)
```

```python
#pairgrid gives you more control over the pairplots

#normally, set it equal to a variable
g = sns.PairGrid(iris)

#show distribution plots on the diagonal of the grind
g.map_diag(sns.distplot)

#show scatter plots on the grids above the diagonal
g.map_upper(plt.scatter)

#show kde plots on the grids below the diagonal
g.map_lower(sns.kdeplot)
```

```python
#Fascett Grid
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data = tips, col='time', row='smoker')

#mapping a plot type and seperating the results based off of the column and row
g.map(sns.distplot,'total_bill')
```

```python
g = sns.FacetGrid(data = tips, col='time', row='smoker')

#mapping a plot type and seperating the results based off of the column and row
#if using a scatterplot, you need 2 arguments and you will get an error if you only have 1. 
g.map(plt.scatter,'total_bill','tip')
```
