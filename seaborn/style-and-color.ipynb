```python
import seaborn as sns 
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
import matplotlib.pyplot as plt
```

```python
#seaborn has a set style argument that allows you to set the style for the rest of the plots. 
#'white' shows a white background
#'ticks' shows the ticks on the axis
#darkgrid shows dark background
#whitegrid shows white background with lines in background
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)

#remove specific 'spines' - the spins are the lines around the border
sns.despine()
```

```python
#change the figure size, aspect ration
plt.figure(figsize=(12,3))
sns.countplot(x='sex', data=tips)
```

```python
#scale and contect - allows you to override default parameters and change plot scales 
sns.set_context('notebook')
sns.countplot(x='sex', data=tips)
```

```python
#pallets and colors
sns.lmplot(x='total_bill',y='tip', data=tips, hue='sex', palette='seismic')
```
